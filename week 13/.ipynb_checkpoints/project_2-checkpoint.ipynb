```python
import pandas as pd
name = ["Akanbi Joel", "Delegiwa Olaye", "Martha Ojeh", "Fadiku Ojo","Olaye Tunde", 
        "Michal Jobs","Adegbesan Ani", "Faith Moses", "Juwon Ade", "Edward Lawal"]
year = [16, 12, 20, 15, 11, 30, 10, 22, 12, 9]
benefit = [300, 150, 100, 450, 200, 170, 250, 80, 120, 280]
accrued = []
for i in range(len(name)):
    accrued.append(year[i] * benefit[i])
data = pd.DataFrame({"Name" : name, "Years of Service" : year, "Accumulated Benefits(units)" : benefit, "Accrued Benefits": accrued})
data
```
